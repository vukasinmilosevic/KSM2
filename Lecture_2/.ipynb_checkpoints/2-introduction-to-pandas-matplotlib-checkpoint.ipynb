### Problem 1
plot histograms couple of features from the dataframe

```python
plt.hist(data['fixed acidity'].values, color='purple', bins=50, alpha=0.5, label='fixed acidity')
plt.hist(data['total sulfur dioxide'].values, color='green', bins=50, alpha=0.5, label='total sulfur dioxide')
plt.hist(data['free sulfur dioxide'].values, color='yellow', bins=50, alpha=0.5, label='free acidity')
plt.legend()
pass
```

### Problem 2
Let's classify wines in:
   * good means quality >= 7
   * bad means quality <= 4
   * mediocre means quality five or six
   
Plot histograms of all features for good and bad wines. Which feature is the most discriminative?
   

```python
gqw = data[data['quality'] >= 7]
bqw = data[data['quality'] <= 4]
mqw = data[data['quality'].between(4, 7, inclusive=False)]

for i in gqw.columns:
    plt.hist(gqw[i], color='green', bins=30, label='Good wine')
    plt.hist(bqw[i], color='red', bins=30, label='Bad wine')
    plt.legend()
    plt.title(i)
    plt.show()
    
print ('The most discriminative features between good and bad wines are density and alcohol')
    
```

### Problem 3   
Plot some scatter plots:
   - alcohol vs fixed acidity for good and bad wines, what can we say about good wines?
   - alcohol vs fixed acidity for quality=5 and quality=6
   - alcohol vs density for good and bad wines. What can we conclude?

```python
plt.scatter(gqw['alcohol'], gqw['fixed acidity'], c='g', label='Good wines')
plt.scatter(bqw['alcohol'], bqw['fixed acidity'], c='r', label='Bad wines')
plt.xlabel('alcohol')
plt.ylabel('fixed acidity')
plt.title('Wines alcohol vs fixed acidity')
plt.legend()
plt.show()

plt.scatter(mqw[mqw['quality'] == 5]['alcohol'], mqw[mqw['quality'] == 5]['fixed acidity'], label='Quality = 5')
plt.scatter(mqw[mqw['quality'] == 6]['alcohol'], mqw[mqw['quality'] == 6]['fixed acidity'], label='Quality = 6')
plt.xlabel('alcohol')
plt.ylabel('fixed acidity')
plt.title('Quality 5 and 6 wines alcohol vs fixed acidity')
plt.legend()
plt.show()

plt.scatter(gqw['alcohol'], gqw['density'], c='g', label='Good wines')
plt.scatter(bqw['alcohol'], bqw['density'], c='r', label='Bad wines')
plt.xlabel('alcohol')
plt.ylabel('density')
plt.title('Wines alcohol vs density')
plt.legend()
plt.show()

print ('Good wines generally have higher alcohol percentage and their relation to fixed acidity is more consistent. Bad wines are little bit denser')
```
