### Problem 2
Let's classify wines in:
   * good means quality >= 7
   * bad means quality <= 4
   * mediocre means quality five or six
   
Plot histograms of all features for good and bad wines. Which feature is the most discriminative?
   

```python

good = data[data['quality'] >= 7]
bad = data[data['quality'] <= 4]
ugly = data[(data['quality'] >4) & (data['quality']<7)]
```

```python
plt.hist(good['fixed acidity'].values, label='good', alpha=0.7)
plt.hist(bad['fixed acidity'].values, label='bad', alpha=0.7)
plt.title('histogram fiksne kiselosti za dobra i losa vina')
plt.xlabel('fixed acidity')
plt.legend()
```

```python
plt.hist(good['volatile acidity'].values, label='good', alpha=0.7)
plt.hist(bad['volatile acidity'].values, label='bad', alpha=0.7)
plt.xlabel('volatile acidity')
plt.legend()
```

```python
plt.hist(good['residual sugar'].values, label='good', alpha=0.7)
plt.hist(bad['residual sugar'].values, label='bad', alpha=0.7)
plt.xlabel('residual sugar')
plt.legend()
```

### Problem 3   
Plot some scatter plots:
   - alcohol vs fixed acidity for good and bad wines, what can we say about good wines?
   - alcohol vs fixed acidity for quality=5 and quality=6
   - alcohol vs density for good and bad wines. What can we conclude?

```python
plt.scatter(good['alcohol'], good['fixed acidity'], c='c', alpha=0.8, marker='D', label='good')
plt.scatter(bad['alcohol'], bad['fixed acidity'], c='g', alpha=0.8,marker='X', label='bad')
plt.xlabel('alcohol')
plt.ylabel('fixed acidity')
plt.legend()
```

```python
q5 = data[data['quality'] == 5]
q6 = data[data['quality'] == 6]
plt.scatter(q6['alcohol'], q6['fixed acidity'], c='g', alpha=0.5,marker='X', label='quality=6')
plt.scatter(q5['alcohol'], q5['fixed acidity'], c='c', alpha=0.5, marker='D', label='quality=5')
plt.xlabel('alcohol')
plt.ylabel('fixed acidity')
plt.legend()
```

```python
plt.scatter(good['alcohol'], good['density'], c='c', alpha=0.8, marker='D', label='good')
plt.scatter(bad['alcohol'], bad['density'], c='g', alpha=0.8,marker='X', label='bad')
plt.xlabel('alcohol')
plt.ylabel('density')
plt.legend()
```
